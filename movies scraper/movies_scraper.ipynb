```python
import scrapy
from scrapy.crawler import CrawlerProcess


class ESSpider(scrapy.Spider):
    # Naming the spider is important if you are running more than one spider of
    # this class simultaneously.
    name = "ESS"
    
    # URL(s) to start with.
    start_urls = [
        'https://www.imdb.com/list/ls045252260/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=c1b9c916-c897-45a2-b228-0013b5e4ff8d&pf_rd_r=752GF7Y1XJBN4NFN3AJV&pf_rd_s=right-1&pf_rd_t=60601&pf_rd_i=best-of.top-movies-of-2018&ref_=ls_bo16_lk10',
    ]

    # Use XPath to parse the response we get.
    def parse(self, response):
        
        # Iterate over every <article> element on the page.
        for posting in response.xpath('//div[@class="lister-item mode-detail"]'):
           
            # Yield a dictionary with the values we want.
            yield {
              
                'name': posting.xpath('div[@class="lister-item-content"]/h3/a/text()').extract_first(),
                'year':posting.xpath('div/h3/span[@class="lister-item-year text-muted unbold"]/text()').extract_first(),
                'rating':posting.xpath('div/div/div/span[@class="ipl-rating-star__rating"]/text()').extract_first(),
                'genre':posting.xpath('div/p/span[@class="genre"]/text()').extract_first()
                
                 }
        
       
        

# Tell the script how to run the crawler by passing in settings.


process = CrawlerProcess({
    'FEED_FORMAT': 'json',         # Store data in JSON format.
    'FEED_URI': 'imdb.json',  # Name our storage file.
    'ROBOTSTXT_OBEY': True,
    'USER_AGENT': 'ThinkfulDataScienceBootcamp_tcbriggs (thinkful.com)',
    'AUTOTHROTTLE_ENABLED': True,
    'HTTPCACHE_ENABLED': True,
    'LOG_ENABLED': False           # Turn off logging for now.
})

# Start the crawler with our spider.
process.crawl(ESSpider)
process.start()
print('Success!')
```

## Rotten Tomatoes

```python
import scrapy
from scrapy.crawler import CrawlerProcess


class ESSpider(scrapy.Spider):
    # Naming the spider is important if you are running more than one spider of
    # this class simultaneously.
    name = "ESS"
    
    # URL(s) to start with.
    start_urls = [
        'https://editorial.rottentomatoes.com/guide/best-movies-of-2019/',
        'https://editorial.rottentomatoes.com/guide/best-movies-of-2018/',
    ]

    # Use XPath to parse the response we get.
    def parse(self, response):
        
        # Iterate over every <article> element on the page.
        for posting in response.xpath('//div[@class="row countdown-item"]/div[@class="col-sm-18 col-full-xs countdown-item-content"]'):
           
            # Yield a dictionary with the values we want.
            yield {
              
                'name': posting.xpath('div[@class="row countdown-item-title-bar"]/div[@class="col-sm-20 col-full-xs"]/div[@class="article_movie_title"]/div/h2/a/text()').extract_first(),
                'year':posting.xpath('div[@class="row countdown-item-title-bar"]/div[@class="col-sm-20 col-full-xs"]/div[@class="article_movie_title"]/div/h2/span[@class="subtle start-year"]/text()').extract_first(),
                'score':posting.xpath('div[@class="row countdown-item-title-bar"]/div[@class="col-sm-20 col-full-xs"]/div[@class="article_movie_title"]/div/h2/span[@class="tMeterScore"]/text()').extract_first(),
                'director':posting.xpath('div[@class="row countdown-item-details"]/div[@class="col-sm-24"]/div[@class="info director"]/a/text()').extract_first(),
                'critics_consensus':posting.xpath('div[@class="row countdown-item-details"]/div[@class="col-sm-24"]/div[@class="info critics-consensus"]/text()').extract_first(),
                
                 }
        
       
        

# Tell the script how to run the crawler by passing in settings.


process = CrawlerProcess({
    'FEED_FORMAT': 'json',         # Store data in JSON format.
    'FEED_URI': 'tomatoes.json',  # Name our storage file.
    'ROBOTSTXT_OBEY': True,
    'USER_AGENT': 'ThinkfulDataScienceBootcamp_tcbriggs (thinkful.com)',
    'AUTOTHROTTLE_ENABLED': True,
    'HTTPCACHE_ENABLED': True,
    'LOG_ENABLED': False           # Turn off logging for now.
})

# Start the crawler with our spider.
process.crawl(ESSpider)
process.start()
print('Success!')
```
